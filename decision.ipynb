### 4. Model Architecture

#### Splitting the dataset

```python
# Splitting the dataset into training and testing dataset, with testing data being 40 %.
Input_train, Input_test, Output_train, Output_test = train_test_split(Input, Output, test_size=0.4, random_state=17)
```

#### Train model

```python
# Fit a Decision Tree Classifier and training the model on Iris dataset.
classifier = tree.DecisionTreeClassifier(random_state=17)
classifier = classifier.fit(Input_train, Output_train)
```

<a id="eval"></a>
### 5. Model Evaluation

#### Evaluation Metrics

```python
# Evaluating the model by evaluating on the testing data.

output_predictions = classifier.predict(Input_test)
print(classification_report(Output_test, output_predictions, target_names=Iris.target_names))
print('\nAccuracy: {0:.4f}'.format(accuracy_score(Output_test, output_predictions)))
```

#### Observations:
The accuracy of the model is 95%.

**F1 score** is a weighted harmonic mean of precision and recall, scores lies in the range [0,1]. Values closer to 1 are better.<br>
    &nbsp; F1 Score = 2*(Recall * Precision) / (Recall + Precision)

**Precision** is the accuracy of positive predictions, high value of precission for each class indicate a good fit.<br>
    &nbsp; Precision = TP/(TP + FP)
    
**Recall** is the fraction of positives that were correctly identified, high value indicates a good fit.<br>
    &nbsp; Recall = TP/(TP+FN)
    
Here,<br>
TP - True Positive (Actually positive and predicted positive).<br>
FP - False Positive (Actually negative but predicted positive).<br>
TN - True Negative (Actually negative and predicted negative).<br>
FN - False Negative (Actually positive but predicted negative).
